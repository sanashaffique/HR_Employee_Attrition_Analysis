```python
#import library
import pandas as pd
from sqlalchemy import create_engine
```

```python
#load dataset
data=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv',index_col='EmployeeNumber')
# view the first 5 rows
pd.set_option('display.max_columns', None)
print(data.head())
```

```python
#dataset info
data.info()
#Observation
"""HR dataset has 1470 rows and 34 columns.There are no non_null entries in dataset.Each field name is properly written.
Each field has correct dtype. Few fields are ordinal in nature ."""
```

```python
#check null values 
print(data.isnull().sum())
#Observation
'''Each column has non-null values.'''
```

```python
#check duplicate
print(data.duplicated().value_counts())
#Observation
'''There is no duplicate value in data.'''
```

```python
# check unique values each column
print(data['EmployeeCount'].unique(),
data['BusinessTravel'].unique(),
data['StandardHours'].unique(),
data['JobInvolvement'].unique(),
data['YearsWithCurrManager'].unique())
```

```python
# remove unvcessary columns
data.drop(columns=['JobInvolvement','YearsWithCurrManager','EmployeeCount','StandardHours','StockOptionLevel'],inplace=True)
```

```python
#recheck columns
data.columns
```

```python
#Modify Col Values 'Over18' 
data['Over18']=data['Over18'].map({'Y':'Yes','N': 'No'})
```

# Convert ALL Ordineal Categories into Labels
'''Attrition
Education
JobSatisfaction
PerformanceRating
RelationshipSatisfaction'''

```python
#Convert Categorical Column into Binary Columns
data['Attrition']=data['Attrition'].map({"Yes":1,"No": 0}).astype(int)
```

```python
#Convert Numeric Column into Categorical Column
data['Education']=data['Education'].map({1: 'Below College',2:'College',3: 'Bachelor',4: 'Master',5 :'Doctor'}).astype(object)
data['JobSatisfaction']=data['JobSatisfaction'].map({1 :'Low',2 :'Medium',3: 'High',4: 'Very High'}).astype(object)
```

```python
# drop irrelevent column 'EnvironmentSatisfaction'
data.drop(columns='EnvironmentSatisfaction',inplace=True)
```

```python
#Convert Numeric Values to Categorical Labels
data['PerformanceRating'] = data['PerformanceRating'].map({
    1: 'Low',
    2: 'Good',
    3: 'Excellent',
    4: 'Outstanding'
}).astype(object)
```

```python
# convert Numeric Values to Categorical Labels
data['RelationshipSatisfaction'] = data['RelationshipSatisfaction'].map({
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}).astype(object)
```

```python
# Recheck All Data
data.info()
# get first 5 rows
print(data.head())
```

```python
#save Cleaned_hr_data
data.to_csv('cleaned_hr_data.csv',index=True)
```
